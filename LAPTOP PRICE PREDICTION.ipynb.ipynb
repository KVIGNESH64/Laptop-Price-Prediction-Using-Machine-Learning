```python
# Feature Engineering and Selection
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

```

```python
#Drop unnecessary columns
df.drop(columns=["Unnamed: 0.1", "Unnamed: 0"], errors='ignore', inplace=True)
df.dropna(inplace=True)
```

```python
df
```

```python
#Filter out invalid values
df = df[df['Weight'].str.contains(r'^\d+(\.\d+)?kg$', na=False)]
df = df[df['Inches'].str.contains(r'^\d+(\.\d+)?$', na=False)]
```

```python
df
```

```python
df['Weight'] = df['Weight'].astype(str)
# remove kgs in Weight by using --replace
df['Weight'] = df['Weight'].str.replace('kg', '').astype(float)
df['Inches'] = df['Inches'].astype(float)
```

```python
df['Weight']
```

```python
df['Inches']
```

```python
# Simplify CPU
# Check if 'Cpu' column exists before proceeding
if 'Cpu' in df.columns:
    df['Cpu_Brand'] = df['Cpu'].apply(lambda x: x.split()[0] + ' ' + x.split()[1])
    df['Cpu_Brand'] = df['Cpu_Brand'].replace({
        'Intel Core': 'Intel',
        'Intel Pentium': 'Intel',
        'Intel Celeron': 'Intel',
        'AMD Ryzen': 'AMD',
        'AMD A9': 'AMD',
        'AMD A6': 'AMD',
        'AMD A4': 'AMD',
        'AMD E2': 'AMD',
        'AMD Athlon': 'AMD'
    })
else:
    print("Column 'Cpu' not found in the DataFrame.")
    # Handle the missing column, e.g., by creating it or skipping this step
```

```python
import pandas as pd

# Assuming df is your DataFrame

# List of potential column names containing memory information
memory_cols = ['Memory', 'Memory1', 'Storage', 'Disc', 'Hard Drive']

# Find the actual column name
memory_col = None
for col in memory_cols:
    if col in df.columns:
        memory_col = col
        break

# If a suitable column is found, extract HDD and SSD sizes
if memory_col:
    df['HDD'] = df[memory_col].apply(lambda x: 1024 if '1TB HDD' in str(x) else 0)  # Convert x to string to avoid errors with non-string values
    df['SSD'] = df[memory_col].apply(lambda x:
        1024 if '1TB SSD' in str(x) else
        512 if '512GB SSD' in str(x) else
        256 if '256GB SSD' in str(x) else
        128 if '128GB SSD' in str(x) else 0)
else:
    print("No suitable memory column found in the DataFrame.")
    # Handle the case where no suitable column is found, e.g., by skipping this step or raising an exception
```

```python
# Drop unused columns, only if they exist
columns_to_drop = ['ScreenResolution', 'Cpu', 'Memory', 'X_res', 'Y_res']
for column in columns_to_drop:
    if column in df.columns:
        df.drop(columns=[column], inplace=True)
```

```python
columns_to_drop
```

```python
df
```

```python
# Encode Categorical Columns
label_encoders = {}
for col in df.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
# Prepare features and target
X = df.drop('Price', axis=1)
y = df['Price']
```

```python
X
```

```python
# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Model Fitting
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict and Evaluate
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("R² Score:", r2)
print("Root Mean Squared Error (RMSE):", rmse)
```

**K-Nearest Neighbor Algorithm**

```python
from sklearn.neighbors import KNeighborsRegressor

```

```python
# Train the KNN model
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(X_train, y_train)
```

```python
# Predict on test set
y_knn_pred = knn_model.predict(X_test)
```

```python
# Evaluate
r2_knn = r2_score(y_test, y_knn_pred)
rmse_knn = np.sqrt(mean_squared_error(y_test, y_knn_pred))

print("KNN R² Score:", r2_knn)
print("KNN Root Mean Squared Error (RMSE):", rmse_knn)
```
